```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
df = pd.read_csv("sequences.csv")
df.head()
```

```python
df = df[df.ID != 'd1erza_']
df = df[df.ID != 'd1ae9a_']
df = df[df.ID != 'd1f44a2']
df = df[df.ID != 'd1g9ng_']
df = df[df.ID != 'd1nos__']
df = df[df.ID != 'd1alq__']
df = df[df.ID != 'd1cc1l_']
df = df[df.ID != 'd1e3db_']
df = df[df.ID != 'd1opc__']






```

```python
superfamilies = np.unique(df.loc[:, 'Superfamily'])
len(superfamilies)
```

```python
labels = pd.DataFrame(index = df.loc[:,'ID'])
labels.head()
```

```python
proteins = df.loc[:,['ID', 'Superfamily']]
```

```python
proteins.head()
```

```python
for column in tqdm(proteins.itertuples()):
    col = []
    for row in proteins.itertuples():
        if column[1] == row[1]:
            col.append(np.NaN)
        else:
            if column[2] == row[2]:
                col.append(1)
            else:
                col.append(0)
    labels[column[1]] = col
labels.head()
```

```python
labels.reset_index()
```

```python
labels.to_csv('trainingLabels.csv')
```
